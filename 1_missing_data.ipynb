```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/weather_data.txt',parse_dates=['day'])
```

```python
df.head()
```

```python
#count missing values in a specfic column
df.temperature.isnull().sum()
#df[df.temperature.isnull()]
#df.isnull().sum()
```

```python
#counting missing values in each column
df.isnull().sum()
df[df.temperature.isnull()]
```

```python
#if we want to use day column as index 
df.set_index('day',inplace=True)
```

```python
df
```

```python
new_df = df.fillna(4)
new_df
```

```python
df.temperature.mean()
```

```python
#fill all nan values with a single value ,not a good solution
#so pandas provides an alternative for that
import numpy as np
new_df = df.fillna({
    'temperature':df.temperature.mean(),
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
#another method
new_df = df.fillna(method='ffill')
# ffill:means carry forward previous value

new_df = df.fillna(method='bfill')
#bfill:means carry forward next value
new_df
```

```python
#filling  values with limited copying from previous or next values

new_df = df.fillna(method='ffill',limit=1) # copied only one time
new_df
```

```python
new_df = df.interpolate()   # fill intermediate values
# default it uses linear interpolation
new_df   
```

```python
# drop a row having nan values
new_df = df.dropna()
new_df
```

```python
# drop a row having all columns are nan values
new_df = df.dropna(how='all')

new_df
```

```python
new_df = df.dropna(thresh=2) #drop all rows having 2 nan values
new_df
```
