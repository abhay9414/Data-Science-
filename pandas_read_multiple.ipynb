Excel files can have multiple sheets.To read a specific sheet or all of them,pandas provides read_excel method which takes an argument called sheetname that tells pandas which sheet to read in the data from. For this, we can either use the sheet name or the sheet number. Sheet numbers start with zero.

If the sheetname argument is not given, it defaults to zero and pandas will import the first sheet.

```python
sheet1 = pd.read_excel(fname,sheetname=0)
s1 = pd.read_excel(fname,sheetname='1900s')
```

```python
s1.head()
```
